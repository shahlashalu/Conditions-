**CONDITIONS**

Used to decide the path of the execution.

```python
#if
#else
#elif
```

```python
age = int(input('Enter the age : '))
if age>=50:
  print('you are eligible for vaccination')
else:
  print('you are not eligible for vaccination')
```

```python
num = int(input('enter the number'))

if num>0:
  print('positive')
elif num<0:
  print('negative')
else:
  print('zero')
```

```python
age = int(input('enter the age : '))
if age<10:   #false
  print('lp')
elif age<13:   #true
  print('up')
elif age<15:
  print('high school')
elif age<17:
  print('hs')
else:
  print('invalid')
```

```python
age = int(input('enter the number : '))
health = input('enter your health condition 1 for critical and 2 for normal.')


if age>60 and health=='2':
  print('you are eligible for second dose')

elif age>60 and health=='1':
  print('you are not eligible for second dose')

elif age>18 and health=='2':
  print('you are eligible for first dose')

elif age>18 and health=='1':
  print('you are not eligible for first dose')

else:
  print('you are not eligible')
```
